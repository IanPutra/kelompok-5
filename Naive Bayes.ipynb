```python
# proses case folding = Case folding adalah bentuk paling sederhana namun sangat efektif untuk text preprocessing. Case folding ini memiliki beberapa fungsi diantaranya, mengubah text menjadi lowercase, menghapus angka, menghapus tanda baca, dan menghapus white space. 
import re

def casefolding(content):
    content = content.lower()
    content = content.strip(" ")
    content = re.sub(r'[?|$|.|!2_:")(-+,]','', content)
    return content
data['content'] = data['content'].apply(casefolding)
data.head(100)
```

```python
# proses tokenizing = tokenizing adalah metode pemisahan kata dalam kalimat. Pemisahan kalimat ini bertujuan untuk analisis teks secara lebih mendalam. 
def token(content):
    nstr = content.split(' ')
    dat= []
    a = -1
    for hu in nstr:
        a = a + 1
    if hu == '':
        dat.append(a)
    p = 0
    b = 0
    for q in dat:
        b = q - p
        del nstr[b]
        p = p + 1
    return nstr
data['content'] = data['content'].apply(token)
data.head(10)
```

```python
data_clean = data_clean.astype({'Label': 'category'})
data_clean = data_clean.astype({'content': 'string'})
data_clean.dtypes
```

```python
#proses TF-IDF adalah algoritma yang berfungsi untuk menganalisa hubungan antara frase atau kalimat dalam sekumpulan dokumen
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf= TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['content'].astype('U'))
text_tf
```

```python
#splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_tf, data_clean['Label'], test_size=0.2, random_state = 42)
```

```python
# performa algoritma naive bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)
print("MultinomialNB Accuracy : ", accuracy_score(y_test,predicted))
print("MultinomialNB Precision: ", precision_score(y_test,predicted, average="binary", pos_label="Negatif"))
print("MultinomialNB Recall   : ", recall_score(y_test,predicted, average="binary", pos_label="Negatif"))
print("MultinomialNB f1_score : ", f1_score(y_test,predicted, average="binary", pos_label="Negatif"))

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('====================================================\n')
print(classification_report(y_test, predicted, zero_division=0))
```
